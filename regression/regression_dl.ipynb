## **Section 3: Data Cleaning & Preprocessing**

This section applies a systematic cleaning strategy so the data becomes suitable for deep learning. Deep learning models are sensitive to scale, missing values, and noisy extreme values, so preprocessing must be done carefully.

### **Cleaning Strategy (Summary Table)**

| Issue | Technique | Rationale |
|---|---|---|
| Duplicates | `drop_duplicates()` | Prevents duplicated signals and leakage |
| >50% missing columns | Drop entirely | Imputation becomes unreliable beyond this threshold |
| Near-zero variance | Drop | No predictive signal and may destabilize scaling |
| Numeric missing values | KNN Imputation | Preserves inter-feature relationships |
| Categorical missing | Mode Imputation | Simple and stable for nominal data |
| Outliers | IQR Winsorization | Caps extreme values without removing rows |
| Low-cardinality categoricals | Label / One-Hot Encoding | Suitable for binary and small nominal features |
| High-cardinality categoricals | Frequency Encoding | Avoids dimensional explosion |

### **Deduplication**

Exact duplicated rows are removed. This prevents the same observation from appearing in both training and testing subsets, which could produce overly optimistic evaluation metrics.

```python
df = df_raw.copy()
del df_raw
gc.collect()

n_before = len(df)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"🗑️  Duplicates removed : {n_before - len(df)} rows ({n_before} → {len(df)})")
```

### **Column Classification and Pruning**

Columns with too many missing values are removed because the majority of their information would need to be artificially reconstructed. Near-zero-variance numeric columns are also removed because they do not contribute meaningful predictive signal.

```python
# Drop columns with >50% missing values, excluding target
threshold = 0.50
high_missing = [c for c in df.columns if df[c].isnull().mean() > threshold and c != TARGET]
if high_missing:
    df.drop(columns=high_missing, inplace=True)
    print(f"🧹 Dropped high-missing columns: {high_missing}")
else:
    print("✅ No high-missing columns dropped.")

num_cols_now = df.select_dtypes(include=np.number).columns.tolist()
cat_cols_now = df.select_dtypes(include=["object", "category"]).columns.tolist()

low_var = [c for c in num_cols_now if c != TARGET and df[c].std() < 1e-6]
if low_var:
    df.drop(columns=low_var, inplace=True)
    print(f"🧹 Dropped near-zero variance columns: {low_var}")
else:
    print("✅ No near-zero variance columns dropped.")

num_cols_now = df.select_dtypes(include=np.number).columns.tolist()
cat_cols_now = df.select_dtypes(include=["object", "category"]).columns.tolist()
feature_cols_num = [c for c in num_cols_now if c != TARGET]
feature_cols_cat = [c for c in cat_cols_now if c != TARGET]

print(f"\n📋 Feature summary:")
print(f"   Numeric features      : {len(feature_cols_num)}")
print(f"   Categorical features  : {len(feature_cols_cat)}")
print(f"   Target column         : {TARGET}")
```

### **Missing Value Imputation**

Numeric missing values are imputed using KNN imputation when possible because it can preserve relationships between numeric features. Categorical missing values are filled using the most frequent category.

```python
# Ensure target is numeric for regression
if not np.issubdtype(df[TARGET].dtype, np.number):
    df[TARGET] = pd.to_numeric(df[TARGET], errors="coerce")

# Remove rows where target is missing, because supervised learning needs a label
n_before = len(df)
df = df.dropna(subset=[TARGET]).reset_index(drop=True)
print(f"🎯 Rows with missing target removed: {n_before - len(df)}")

# Refresh columns after possible conversion
feature_cols_num = [c for c in df.select_dtypes(include=np.number).columns.tolist() if c != TARGET]
feature_cols_cat = [c for c in df.select_dtypes(include=["object", "category"]).columns.tolist() if c != TARGET]

if feature_cols_num and df[feature_cols_num].isnull().any().any():
    n_neighbors = min(5, max(1, len(df) - 1))
    knn_imp = KNNImputer(n_neighbors=n_neighbors)
    df[feature_cols_num] = knn_imp.fit_transform(df[feature_cols_num])
    print(f"✅ Numeric missing values imputed using KNNImputer(k={n_neighbors}).")
else:
    print("✅ No numeric missing values to impute.")

if feature_cols_cat and df[feature_cols_cat].isnull().any().any():
    mode_imp = SimpleImputer(strategy="most_frequent")
    df[feature_cols_cat] = mode_imp.fit_transform(df[feature_cols_cat])
    print("✅ Categorical missing values imputed using mode imputation.")
else:
    print("✅ No categorical missing values to impute.")

print(f"\nRemaining nulls: {df.isnull().sum().sum()}")
```

### **Outlier Treatment via Winsorization**

Outliers are treated using IQR-based winsorization. Instead of deleting rows, extreme values are clipped to a lower and upper bound. This keeps the data size stable while reducing the influence of extreme observations.

```python
def winsorize_iqr(series: pd.Series, factor: float = 1.5) -> pd.Series:
    Q1, Q3 = series.quantile(0.25), series.quantile(0.75)
    IQR = Q3 - Q1
    if IQR == 0 or pd.isna(IQR):
        return series
    return series.clip(lower=Q1 - factor * IQR, upper=Q3 + factor * IQR)

for col in feature_cols_num:
    df[col] = winsorize_iqr(df[col])

print(f"✅ IQR Winsorization applied to {len(feature_cols_num)} numeric feature columns.")
```

### **Categorical Encoding**

Deep learning models require numeric tensors, so categorical features must be encoded. Binary categorical columns use label encoding, low-cardinality columns use one-hot encoding, and high-cardinality columns use frequency encoding to prevent too many sparse features.

```python
def frequency_encode(df: pd.DataFrame, col: str) -> pd.DataFrame:
    freq = df[col].value_counts(normalize=True)
    df[col + "_freq"] = df[col].map(freq).astype(float)
    df.drop(columns=[col], inplace=True)
    return df

for col in list(feature_cols_cat):
    n_unique = df[col].nunique(dropna=True)
    if n_unique <= 1:
        df.drop(columns=[col], inplace=True)
        print(f"Dropped categorical column with <=1 unique value: {col}")
    elif n_unique == 2:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
    elif n_unique <= 10:
        dummies = pd.get_dummies(df[col].astype(str), prefix=col, drop_first=True, dtype=int)
        df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
    else:
        df = frequency_encode(df, col)

print(f"✅ Encoding complete. Shape: {df.shape}")
display(df.head())
```

## **Section 4: Feature Engineering & Selection**

This section creates additional numeric signals and then selects the most relevant features. Feature selection is important because deep learning models can overfit when the number of features is unnecessarily large relative to the dataset size.

### **Feature Engineering**

The notebook creates squared terms and interaction terms from the top correlated features. It also applies `log1p` transformations to right-skewed non-negative features. These transformations help the neural network learn nonlinear patterns more easily.

```python
X_all = df.drop(columns=[TARGET]).copy()
y_all = df[TARGET].copy()

# Keep only numeric features after encoding
X_all = X_all.select_dtypes(include=np.number)

# Basic cleanup for infinite values
X_all = X_all.replace([np.inf, -np.inf], np.nan).fillna(0)
y_all = y_all.replace([np.inf, -np.inf], np.nan).fillna(y_all.median())

corr_with_target = X_all.corrwith(y_all).replace([np.inf, -np.inf], np.nan).fillna(0).abs()
top5 = corr_with_target.nlargest(min(5, len(corr_with_target))).index.tolist()

print("🔝 Top features by |Pearson| correlation with target:")
for f in top5:
    print(f"   {f}: {corr_with_target[f]:.4f}")

engineered = []
for i, col in enumerate(top5):
    X_all[f"{col}_sq"] = X_all[col] ** 2
    engineered.append(f"{col}_sq")

    if i + 1 < len(top5):
        col2 = top5[i + 1]
        X_all[f"{col}_x_{col2}"] = X_all[col] * X_all[col2]
        engineered.append(f"{col}_x_{col2}")

for col in list(X_all.select_dtypes(include=np.number).columns):
    if X_all[col].min() >= 0 and X_all[col].skew() > 0.75:
        X_all[f"{col}_log1p"] = np.log1p(X_all[col])
        engineered.append(f"{col}_log1p")

X_all = X_all.replace([np.inf, -np.inf], np.nan).fillna(0)
print(f"\n🔧 Engineered {len(engineered)} new features.")
print(f"   Feature space: {X_all.shape[1]} total features")
```

### **Feature Selection via Combined Ranking**

Two feature selection signals are combined:

1. `mutual_info_regression` captures nonlinear relationships between features and target.
2. `ExtraTreesRegressor` estimates feature importance using an ensemble of randomized trees.

The final selected features are taken from the best combined ranking.

```python
K = min(30, X_all.shape[1])

mi_scores = mutual_info_regression(X_all.fillna(0), y_all, random_state=SEED)
mi_series = pd.Series(mi_scores, index=X_all.columns).sort_values(ascending=False)

et = ExtraTreesRegressor(n_estimators=100, random_state=SEED, n_jobs=-1)
et.fit(X_all.fillna(0), y_all)
et_imp = pd.Series(et.feature_importances_, index=X_all.columns).sort_values(ascending=False)

rank_mi = mi_series.rank(ascending=False)
rank_et = et_imp.rank(ascending=False)
combined = (rank_mi + rank_et).sort_values()

selected_features = combined.head(K).index.tolist()
print(f"✅ Selected {K} features by combined MI + ExtraTrees rank.")

plt.figure(figsize=(10, 7))
mi_series.head(20).sort_values().plot.barh()
plt.title("Top-20 Mutual Information Scores", fontsize=12)
plt.xlabel("Mutual Information")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 7))
et_imp.head(20).sort_values().plot.barh()
plt.title("Top-20 ExtraTrees Feature Importance", fontsize=12)
plt.xlabel("Feature Importance")
plt.tight_layout()
plt.show()
```

## **Section 5: Train / Validation / Test Split and Scaling**

The dataset is split into training, validation, and testing subsets using a 70/15/15 ratio. Scaling is mandatory for deep learning because neural networks train more stably when input features and target values are normalized.

### **Split and Scaling Strategy**

- `RobustScaler` is used for input features because it is less sensitive to remaining outliers.
- `StandardScaler` is used for the regression target because neural networks optimize more smoothly when the target has approximately zero mean and unit variance.
- The test set is never used for fitting scalers, oversampling, tuning, or early stopping.

```python
X = X_all[selected_features].values.astype(np.float32)
y = y_all.values.astype(np.float32)

# Free memory from intermediate dataframe objects
try:
    del df, X_all, y_all
    gc.collect()
except Exception:
    pass

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y, test_size=0.15, random_state=SEED
)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=0.15/(1-0.15), random_state=SEED
)

feature_scaler = RobustScaler()
X_train_s = feature_scaler.fit_transform(X_train).astype(np.float32)
X_val_s = feature_scaler.transform(X_val).astype(np.float32)
X_test_s = feature_scaler.transform(X_test).astype(np.float32)

print("Dataset split summary:")
print(f"  Train      : {X_train_s.shape[0]:>6,} rows")
print(f"  Validation : {X_val_s.shape[0]:>6,} rows")
print(f"  Test       : {X_test_s.shape[0]:>6,} rows")
print(f"  Features   : {X_train_s.shape[1]}")
```

## **Section 6: Regression Target Imbalance Handling with SMOTE-style Oversampling**

Classic class imbalance handling does not directly apply to regression because the target is continuous. However, regression datasets can still be imbalanced when certain target ranges are rare, for example very low or very high target values.

To address this, the continuous target is divided into bins. Underrepresented bins are then oversampled using interpolation between existing samples inside the same bin. This creates synthetic training examples while keeping validation and test data untouched.

### **SMOTER-lite Implementation**

The function below is a lightweight SMOTE-style method for regression:

1. Divide continuous target values into equal-width bins.
2. Count the number of samples in each target range.
3. Identify underrepresented bins.
4. Generate synthetic rows by interpolating two randomly selected samples from the same bin.
5. Apply the procedure only to the training set to avoid leakage.

```python
def smoter_lite_resample(
    X_train,
    y_train,
    n_bins=10,
    min_bin_size=2,
    max_multiplier=2.0,
    random_state=SEED,
):
    '''SMOTE-style oversampling for regression target imbalance.

    This function balances rare target ranges by interpolating features and target values
    between samples that belong to the same target bin.
    '''
    rng = np.random.default_rng(random_state)
    X_train = np.asarray(X_train, dtype=np.float32)
    y_train = np.asarray(y_train, dtype=np.float32)

    # Equal-width bins expose target-range imbalance better than quantile bins.
    bins = pd.cut(y_train, bins=n_bins, labels=False, duplicates="drop", include_lowest=True)
    bins = pd.Series(bins).fillna(0).astype(int).values

    unique_bins, counts = np.unique(bins, return_counts=True)
    count_map_before = dict(zip(unique_bins, counts))

    nonzero_counts = counts[counts > 0]
    if len(nonzero_counts) == 0:
        print("⚠️  No valid bins found. Returning original training data.")
        return X_train, y_train, count_map_before, count_map_before

    median_count = int(np.median(nonzero_counts))
    max_count = int(np.max(nonzero_counts))
    target_count = max(median_count, int(np.ceil(median_count * max_multiplier)))
    target_count = min(target_count, max_count)

    X_synth, y_synth = [], []

    for b in unique_bins:
        idx = np.where(bins == b)[0]
        n_current = len(idx)

        if n_current < min_bin_size:
            # Too few samples to interpolate safely.
            continue

        n_to_generate = max(0, target_count - n_current)
        for _ in range(n_to_generate):
            i, j = rng.choice(idx, size=2, replace=True)
            lam = rng.random()
            x_new = X_train[i] + lam * (X_train[j] - X_train[i])
            y_new = y_train[i] + lam * (y_train[j] - y_train[i])
            X_synth.append(x_new)
            y_synth.append(y_new)

    if len(X_synth) == 0:
        print("ℹ️  No synthetic samples generated. Target bins are already reasonably balanced.")
        count_map_after = count_map_before
        return X_train, y_train, count_map_before, count_map_after

    X_res = np.vstack([X_train, np.asarray(X_synth, dtype=np.float32)])
    y_res = np.concatenate([y_train, np.asarray(y_synth, dtype=np.float32)])

    bins_after = pd.cut(y_res, bins=n_bins, labels=False, duplicates="drop", include_lowest=True)
    bins_after = pd.Series(bins_after).fillna(0).astype(int).values
    unique_after, counts_after = np.unique(bins_after, return_counts=True)
    count_map_after = dict(zip(unique_after, counts_after))

    return X_res.astype(np.float32), y_res.astype(np.float32), count_map_before, count_map_after


X_train_bal, y_train_bal, bin_before, bin_after = smoter_lite_resample(
    X_train_s, y_train, n_bins=10, max_multiplier=2.0, random_state=SEED
)

print("📊 Target-bin distribution before SMOTE-style oversampling:")
print(bin_before)
print("\n📊 Target-bin distribution after SMOTE-style oversampling:")
print(bin_after)
print(f"\nRows before: {len(X_train_s):,} | Rows after: {len(X_train_bal):,}")

plt.figure(figsize=(10, 4))
plt.hist(y_train, bins=30, alpha=0.6, label="Original train target")
plt.hist(y_train_bal, bins=30, alpha=0.6, label="SMOTE-style train target")
plt.title("Target Distribution Before vs After SMOTE-style Oversampling")
plt.xlabel(TARGET)
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

### **Evaluation Helper and Plotting Utilities**

The evaluation function computes core regression metrics:

- **MSE**: average squared error.
- **RMSE**: square root of MSE and easier to interpret because it has the same unit as the target.
- **MAE**: average absolute prediction error.
- **R²**: proportion of target variance explained by the model.
- **MAPE**: percentage error, calculated safely to avoid division by zero.

```python
def inverse_target(y_scaled):
    y_scaled = np.asarray(y_scaled).reshape(-1, 1)
    return target_scaler.inverse_transform(y_scaled).ravel()


def predict_original_scale(model, X_data, batch_size=1024):
    y_pred_scaled = model.predict(X_data, batch_size=batch_size, verbose=0).ravel()
    return inverse_target(y_pred_scaled)


def safe_mape(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.abs(y_true) > 1e-8
    if not np.any(mask):
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def evaluate(y_true, y_pred, prefix=""):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = safe_mape(y_true, y_pred)
    return {
        f"{prefix}MSE": float(mse),
        f"{prefix}RMSE": float(rmse),
        f"{prefix}MAE": float(mae),
        f"{prefix}R2": float(r2),
        f"{prefix}MAPE": float(mape) if not np.isnan(mape) else np.nan,
    }


def plot_predictions(y_true, y_pred, title="Actual vs Predicted"):
    residuals = y_true - y_pred

    plt.figure(figsize=(6, 5))
    mn, mx = min(y_true.min(), y_pred.min()), max(y_true.max(), y_pred.max())
    plt.scatter(y_true, y_pred, alpha=0.45, s=18)
    plt.plot([mn, mx], [mn, mx], "--", linewidth=1.5)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"{title}\nActual vs Predicted")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.hist(residuals, bins=40, edgecolor="white", alpha=0.85)
    plt.axvline(0, linewidth=1.5, linestyle="--")
    plt.title(f"{title}\nResidual Distribution")
    plt.xlabel("Residual")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.scatter(y_pred, residuals, alpha=0.45, s=18)
    plt.axhline(0, linewidth=1.5, linestyle="--")
    plt.xlabel("Predicted")
    plt.ylabel("Residual")
    plt.title(f"{title}\nResidual vs Predicted")
    plt.tight_layout()
    plt.show()


def plot_training_history(history, title="Training History"):
    hist = pd.DataFrame(history.history)

    plt.figure(figsize=(8, 4))
    plt.plot(hist["loss"], label="train_loss")
    if "val_loss" in hist.columns:
        plt.plot(hist["val_loss"], label="val_loss")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## **Section 8: Deep Learning Model Architecture**

This notebook uses a Multi-Layer Perceptron (MLP) because the dataset is tabular. The architecture combines Dense layers, Batch Normalization, activation functions, Dropout, and L2 regularization.

### **Reusable Keras MLP Builder**

The model builder accepts hidden layer sizes, dropout rate, learning rate, activation function, and L2 regularization. This makes the same function reusable for baseline experiments and Optuna tuning.

```python
def build_mlp_regressor(
    input_dim,
    hidden_layers=(128, 64),
    dropout_rate=0.20,
    learning_rate=1e-3,
    activation="relu",
    l2_reg=1e-5,
    loss_name="mse",
):
    tf.keras.backend.clear_session()

    inputs = keras.Input(shape=(input_dim,), name="features")
    x = inputs

    for i, units in enumerate(hidden_layers):
        x = layers.Dense(
            units,
            activation=None,
            kernel_regularizer=regularizers.l2(l2_reg),
            name=f"dense_{i+1}",
        )(x)
        x = layers.BatchNormalization(name=f"batch_norm_{i+1}")(x)
        x = layers.Activation(activation, name=f"activation_{i+1}")(x)
        if dropout_rate > 0:
            x = layers.Dropout(dropout_rate, seed=SEED, name=f"dropout_{i+1}")(x)

    # Mixed precision: keep final output as float32 for numerical stability
    outputs = layers.Dense(1, activation="linear", dtype="float32", name="regression_output")(x)
    model = keras.Model(inputs=inputs, outputs=outputs, name="MLP_Regressor")

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    loss = keras.losses.Huber() if loss_name == "huber" else "mse"

    model.compile(
        optimizer=optimizer,
        loss=loss,
        metrics=[
            keras.metrics.MeanAbsoluteError(name="mae"),
            keras.metrics.RootMeanSquaredError(name="rmse"),
        ],
    )
    return model


# ── SPEED OPTIMIZATION 2: tf.data pipeline helper ───────────────────────────
def make_tf_dataset(X, y, batch_size, shuffle=False, seed=SEED):
    """Wrap numpy arrays into a cached + prefetched tf.data.Dataset.
    
    cache() keeps the dataset in memory after the first epoch so subsequent
    epochs skip I/O entirely.  prefetch(AUTOTUNE) overlaps data prep with
    GPU kernel execution, removing the data-starvation gap between batches.
    """
    ds = tf.data.Dataset.from_tensor_slices((X, y))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(X), seed=seed, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).cache().prefetch(tf.data.AUTOTUNE)
    return ds


# ── SPEED OPTIMIZATION 3: Reduced EarlyStopping patience ────────────────────
# patience=25/30 wastes many epochs waiting for improvement that won't come.
# Combined with ReduceLROnPlateau, patience=12 is sufficient to converge.
def get_callbacks(patience=12):
    return [
        keras.callbacks.EarlyStopping(
            monitor="val_loss",
            patience=patience,          # was 25/20/30 — now 12 default
            restore_best_weights=True,
            verbose=0,
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=max(4, patience // 3),
            min_lr=1e-6,
            verbose=0,
        ),
    ]

print("✅ Keras MLP builder + tf.data helper ready.")
```

## **Section 9: Baseline Deep Learning Training with MLFlow Logging**

Three baseline neural network configurations are trained and compared. Each model is logged into MLFlow with its architecture parameters and validation metrics.

### **Baseline Model Registry**

The baseline registry defines three MLP configurations from simple to deeper architecture. These models establish a baseline before Optuna searches for better hyperparameters.

```python
# ── SPEED OPTIMIZATION 4: Larger batch sizes ──────────────────────────────
# Doubling batch size halves the number of gradient steps per epoch,
# which translates directly to faster wall-clock time with minimal
# accuracy impact when combined with a proportionally tuned LR.
baseline_configs = {
    "MLP_Small": {
        "hidden_layers": (128, 64),
        "dropout_rate": 0.10,
        "learning_rate": 1e-3,
        "activation": "relu",
        "l2_reg": 1e-5,
        "batch_size": 256,        # was 128
        "loss_name": "mse",
    },
    "MLP_BatchNorm": {
        "hidden_layers": (256, 128, 64),
        "dropout_rate": 0.20,
        "learning_rate": 7e-4,
        "activation": "relu",
        "l2_reg": 1e-4,
        "batch_size": 256,        # was 128
        "loss_name": "mse",
    },
    "MLP_Deep_Huber": {
        "hidden_layers": (512, 256, 128),
        "dropout_rate": 0.25,
        "learning_rate": 5e-4,
        "activation": "elu",
        "l2_reg": 1e-4,
        "batch_size": 256,        # was 128
        "loss_name": "huber",
    },
}

print("✅ Baseline registry created:")
for name, cfg in baseline_configs.items():
    print(f"   {name}: {cfg}")
```

### **Training Loop with MLFlow Logging**

Each baseline model is trained using early stopping. Validation predictions are inverse-transformed back to the original target scale before metrics are calculated.

```python
if mlflow.active_run():
    mlflow.end_run()

results_ledger = []
trained_models = {}
histories = {}

for model_name, cfg in baseline_configs.items():
    print("\n" + "=" * 80)
    print(f"Training baseline model: {model_name}")
    print("=" * 80)

    model = build_mlp_regressor(
        input_dim=X_train_bal.shape[1],
        hidden_layers=cfg["hidden_layers"],
        dropout_rate=cfg["dropout_rate"],
        learning_rate=cfg["learning_rate"],
        activation=cfg["activation"],
        l2_reg=cfg["l2_reg"],
        loss_name=cfg["loss_name"],
    )

    # ── SPEED OPTIMIZATION 5: Use tf.data pipeline for training ────────────
    train_ds = make_tf_dataset(X_train_bal, y_train_bal_s, cfg["batch_size"], shuffle=True)
    val_ds   = make_tf_dataset(X_val_s,     y_val_s,       cfg["batch_size"], shuffle=False)

    start_time = time.time()
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=200,
        callbacks=get_callbacks(patience=12),   # was patience=25
        verbose=0,
    )
    train_time = time.time() - start_time

    y_pred_val = predict_original_scale(model, X_val_s)
    metrics_val = evaluate(y_val, y_pred_val, prefix="val_")
    metrics_val["train_time_sec"] = float(train_time)
    metrics_val["epochs_trained"] = int(len(history.history["loss"]))

    print(f"{model_name} -> Val R²={metrics_val['val_R2']:+.4f} | RMSE={metrics_val['val_RMSE']:.4f} | MAE={metrics_val['val_MAE']:.4f} | ⏱ {train_time:.1f}s ({metrics_val['epochs_trained']} epochs)")

    try:
        with mlflow.start_run(run_name=model_name):
            mlflow.log_params({
                "model": model_name,
                "hidden_layers": str(cfg["hidden_layers"]),
                "dropout_rate": cfg["dropout_rate"],
                "learning_rate": cfg["learning_rate"],
                "activation": cfg["activation"],
                "l2_reg": cfg["l2_reg"],
                "batch_size": cfg["batch_size"],
                "loss_name": cfg["loss_name"],
                "smote_style_rows_before": len(X_train_s),
                "smote_style_rows_after": len(X_train_bal),
            })
            mlflow.log_metrics(metrics_val)
            try:
                mlflow.tensorflow.log_model(model, artifact_path="model")
                print("   ✅ Keras model artifact logged.")
            except Exception as log_err:
                print(f"   ⚠️  log_model skipped: {log_err}")
    except Exception as run_err:
        print(f"   ⚠️  MLFlow run failed: {run_err}")
        if mlflow.active_run():
            mlflow.end_run(status="FAILED")

    results_ledger.append({"Model": model_name, "Stage": "baseline", **metrics_val})
    trained_models[model_name] = model
    histories[model_name] = history

    plot_training_history(history, title=f"Training History — {model_name}")
    gc.collect()

print("\n✅ Baseline training completed.")
```

## **Section 10: Hyperparameter Tuning with Optuna**

Optuna is used for basic hyperparameter tuning of the MLP architecture. The search objective minimizes validation RMSE in the original target scale.

### **Optuna Objective Function**

The search space includes number of layers, units per layer, dropout rate, learning rate, activation function, L2 regularization, batch size, and loss function. Early stopping is used inside every trial to reduce unnecessary training time.

```python
N_TRIALS = 15

optuna.logging.set_verbosity(optuna.logging.WARNING)

try:
    from optuna.integration import TFKerasPruningCallback
except ImportError:
    try:
        from optuna_integration import TFKerasPruningCallback
    except ImportError:
        TFKerasPruningCallback = None

def objective_mlp(trial):
    n_layers = trial.suggest_int("n_layers", 2, 4)
    hidden_layers = []
    for i in range(n_layers):
        units = trial.suggest_categorical(f"units_l{i+1}", [64, 128, 256, 512])
        hidden_layers.append(units)

    dropout_rate  = trial.suggest_float("dropout_rate", 0.0, 0.40)
    learning_rate = trial.suggest_float("learning_rate", 1e-5, 3e-3, log=True)
    activation    = trial.suggest_categorical("activation", ["relu", "elu", "swish"])
    l2_reg        = trial.suggest_float("l2_reg", 1e-7, 1e-3, log=True)
    batch_size    = trial.suggest_categorical("batch_size", [128, 256, 512])
    loss_name     = trial.suggest_categorical("loss_name", ["mse", "huber"])

    model = build_mlp_regressor(
        input_dim=X_train_bal.shape[1],
        hidden_layers=tuple(hidden_layers),
        dropout_rate=dropout_rate,
        learning_rate=learning_rate,
        activation=activation,
        l2_reg=l2_reg,
        loss_name=loss_name,
    )

    train_ds = make_tf_dataset(X_train_bal, y_train_bal_s, batch_size, shuffle=True)
    val_ds   = make_tf_dataset(X_val_s,     y_val_s,       batch_size, shuffle=False)

    cb_list = get_callbacks(patience=10)
    if TFKerasPruningCallback is not None:
        pruning_cb = TFKerasPruningCallback(trial, "val_loss")
        cb_list.append(pruning_cb)

    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=80,
        callbacks=cb_list,
        verbose=0,
    )

    y_pred_val = predict_original_scale(model, X_val_s)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))

    trial.set_user_attr("epochs_trained", len(history.history["loss"]))
    tf.keras.backend.clear_session()
    gc.collect()
    return rmse


study_mlp = optuna.create_study(
    direction="minimize",
    study_name="mlp_regression_study",
    sampler=optuna.samplers.TPESampler(seed=SEED),
    pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=10),
)
study_mlp.optimize(objective_mlp, n_trials=N_TRIALS, n_jobs=1)

best_mlp_params = study_mlp.best_params
pruned_count   = len([t for t in study_mlp.trials if t.state == optuna.trial.TrialState.PRUNED])
complete_count = len([t for t in study_mlp.trials if t.state == optuna.trial.TrialState.COMPLETE])
print(f"\n✅ Best MLP validation RMSE: {study_mlp.best_value:.4f}")
print(f"   Completed: {complete_count} | Pruned (saved): {pruned_count}")
print(json.dumps(best_mlp_params, indent=2))
```

## **Section 13: Model Interpretation with LIME**

LIME explains individual predictions by approximating the neural network locally with a simpler interpretable model. This is useful because deep learning models are less transparent than linear models or decision trees.

### **Best Model Selection for LIME**

The best model is selected from the leaderboard based on R². LIME then uses a prediction wrapper that transforms the neural network output back to the original target scale.

```python
best_row = leaderboard_df.sort_values("R2", ascending=False).iloc[0]
best_model_name = best_row["Model"]
best_model_obj = trained_models[best_model_name]

print(f"🏆 Best model for LIME: {best_model_name} (R²={best_row['R2']:.4f}, RMSE={best_row['RMSE']:.4f})")


def lime_predict_fn(data):
    data = np.asarray(data, dtype=np.float32)
    return predict_original_scale(best_model_obj, data)

lime_explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=X_train_bal,
    feature_names=selected_features,
    mode="regression",
    discretize_continuous=True,
    random_state=SEED,
)

print("✅ LIME explainer initialized.")
```

### **Per-Instance Explanations**

Three representative test samples are explained:

1. A typical prediction near the median predicted value.
2. The highest predicted value.
3. The lowest predicted value.

This provides insight into how the model behaves across different target ranges.

```python
y_pred_best = lime_predict_fn(X_test_s)

idx_median = int(np.argmin(np.abs(y_pred_best - np.median(y_pred_best))))
idx_max    = int(np.argmax(y_pred_best))
idx_min    = int(np.argmin(y_pred_best))

sample_info = [
    (idx_median, "Typical prediction near median"),
    (idx_max,    "Highest predicted value"),
    (idx_min,    "Lowest predicted value"),
]

lime_weights_all = []

for idx, label in sample_info:
    print("\n" + "─" * 50)
    print(f"🔍 LIME Explanation — {label}")
    print(f"   Actual    : {y_test[idx]:.4f}")
    print(f"   Predicted : {y_pred_best[idx]:.4f}")
    print("─" * 50)

    explanation = lime_explainer.explain_instance(
        data_row=X_test_s[idx],
        predict_fn=lime_predict_fn,
        num_features=min(15, len(selected_features)),
        num_samples=500,        # was 1000 — 500 is statistically sufficient
    )

    lime_weights_all.append(dict(explanation.as_list()))
    feats, weights = zip(*explanation.as_list())

    plt.figure(figsize=(10, max(4, len(feats) * 0.4)))
    y_pos  = range(len(feats))
    colors = ["#2ecc71" if w > 0 else "#e74c3c" for w in weights]
    plt.barh(y_pos, weights, color=colors, edgecolor="white")
    plt.yticks(y_pos, feats, fontsize=9)
    plt.axvline(0, linewidth=1)
    plt.title(f"LIME — {label}\nActual={y_test[idx]:.4f} | Predicted={y_pred_best[idx]:.4f}")
    plt.xlabel("Feature contribution to prediction")
    green_p = mpatches.Patch(color="#2ecc71", label="Positive contribution")
    red_p   = mpatches.Patch(color="#e74c3c", label="Negative contribution")
    plt.legend(handles=[green_p, red_p], fontsize=9)
    plt.tight_layout()
    plt.show()
```

### **Aggregate LIME Importance**

The aggregate LIME view averages absolute feature contributions across multiple test samples. This provides a more stable view of which features repeatedly influence model predictions.

```python
print("🔄 Computing aggregate LIME importance...")
n_lime_samples = min(20, len(X_test_s))
rng     = np.random.default_rng(SEED)
rng_idx = rng.choice(len(X_test_s), n_lime_samples, replace=False)

feature_importance_agg = {f: 0.0 for f in selected_features}

for idx in rng_idx:
    exp = lime_explainer.explain_instance(
        data_row=X_test_s[idx],
        predict_fn=lime_predict_fn,
        num_features=len(selected_features),
        num_samples=300,            # was 500
    )
    for feat, weight in exp.as_list():
        matched = [f for f in selected_features if f in feat]
        key = matched[0] if matched else feat
        feature_importance_agg[key] = feature_importance_agg.get(key, 0) + abs(weight)

lime_agg_series = pd.Series(feature_importance_agg).sort_values(ascending=True) / n_lime_samples
lime_agg_top20  = lime_agg_series.tail(20)

plt.figure(figsize=(10, 7))
lime_agg_top20.plot.barh()
plt.title(f"Aggregate LIME Feature Importance\nMean |weight| over {n_lime_samples} test samples — {best_model_name}")
plt.xlabel("Mean |LIME weight|")
plt.tight_layout()
plt.show()

print("\n🔑 Top-10 most influential features by aggregate LIME:")
print(lime_agg_series.tail(10).sort_values(ascending=False).to_string())
```
